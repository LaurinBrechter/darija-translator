```python
import pandas as pd
import os
```

```python
all_words = []

def append_frame(name, cateogory):
    frame = pd.read_csv('dataset/syntactic categories/' + name + '.csv')
    frame['type'] = cateogory
    all_words.append(frame)

append_frame('adverbs', 'adverb')
append_frame('prepositions', 'preposition')
append_frame('pronouns', 'pronoun')
append_frame('verbs', 'verb')
append_frame('nouns', 'noun')
append_frame('adjectives', 'adjective')

all_words = pd.concat(all_words)
```

```python
all_words.to_csv('all_words.csv', index=False)
```

```python
categories = []

for category in os.listdir('dataset/semantic categories')[:5]:
    category_df = pd.read_csv(f'dataset/semantic categories/{category}')
    category_df['category'] = category.split('.')[0]
    category_df = category_df.drop(columns=['n1', 'n2', 'n3', 'n4'])

    categories.append(category_df)

categories = pd.concat(categories)
```

```python
all_words.merge(categories, on=['darija_ar'], how='inner')
```
